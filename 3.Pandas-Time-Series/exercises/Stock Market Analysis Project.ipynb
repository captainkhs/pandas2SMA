## Daily Percentage Change
첫째로 daily percentage change 를 계산해보겠습니다. Daily percentage change 를 수학식으로 다음과 같이 정의합니다:

$ r_t = \frac{p_t}{p_{t-1}} -1$

이것은 t 시간의 리턴값 $r_t$ 를 t 시간의 가격을($p_t$) t-1 시간의(하루 전날) 가격으로($p_{t-1}$) 나눈 후 1을 뺀 값으로 정의한 것입니다.

기본적으로 이 지표는 주식을 산 후 다음날 팔았을 경우에 대한 percent gain (또는 loss) 를 알려줍니다. 주가를 예측하는 지표는 될 수 없지만 주가가 얼마나 변덕스러운지를(변동지수; volatility) 이 지표를 통해 알 수 있습니다. 매일매일의 이 값의 지표의 분포가 주식의 변동지수가 됩니다. 

daily percent change 를 구한 후 이를 히스토그램으로 시각화 하여 어떤 주식이 가장 안정적인지 판단해봅시다!

## Cumulative Daily Returns

Cumulative Daily Returns 는 시계열 데이터의 시작시점에 $1 를 투자했을 경우 오늘 시점에 얼마가 되었는지를 나타냅니다. 현 시점의 주가와는 달리 매일매일의 수익을 반영합니다. 간단히 계산하기 위해 배당금은(dividend) 고려하지 않습니다. 다음 예시를 참고하세요:

활발히 거래되고 있는 'ABC' 주식이 있습니다. ABC 의 가격은 날짜별로 아래와 같았습니다.

    Date                        Price
    01/01/2018                   10
    01/02/2018                   15
    01/03/2018                   20
    01/04/2018                   25

**Daily Return** : Daily return 은 전날과 비교한 주식의 수익/손해(profit/loss) 입니다. 값이 1 을 초과하면 수익인 경우고, 1 미만이면 손해를 나타냅니다. 퍼센트로 표시하기도 하는데 이경우 0을 기준으로 수익과 손해를 나타내게 됩니다. 위 주가를 기준으로 한 Daily Return 은 아래와 같습니다. 

    Date                         Daily Return                  %Daily Return
    01/01/2018                 10/10 =  1                          -   
    01/02/2018                 15/10 =  3/2                       50%
    01/03/2018                 20/15 =  4/3                       33%
    01/04/2018                 25/20 =  5/4                       20%

**Cumulative Return**: 투자자들은 Daily Return 값을 통해서 특정 시점까지의 이득을 직관적으로 알기는 어렵고, 변동이 심할 수록 더욱 그렇습니다. Cumulative Return 은 투자가 이뤄진 날에 따라 계산됩니다. 이 값이 1보다 크면 수익을 내고 있고 그렇지 않으면 손해가 나고 있는 상태가 됩니다. 위 주가를 기준으로 계산한 Cumulative Return 은 아래와 같습니다.

    Date                       Cumulative Return         %Cumulative Return
    01/01/2018                  10/10 =  1                         100 %   
    01/02/2018                  15/10 =  3/2                       150 %
    01/03/2018                  20/10 =  2                         200 %
    01/04/2018                  25/10 =  5/2                       250 %

cumulative daily return 의 공식은 다음과 같습니다:

$ i_i = (1+r_t) * i_{t-1} $

이전 시점 t-1 에 이뤄진 투자 $i_{t-1}$ 과 percent return $r_t$ 에 1을 더한 값의 곱입니다. 그런데 Pandas 에는 이를 내장함수 cumprod() 메서드로 구현해놓았고 다음과 같이 사용할 수 있습니다:

    df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()
